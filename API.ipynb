```python
import requests
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import json
```

```python
def requests_get(url):
    r = requests.get(url)
    return r
```

```python
def dump_json(name, r):
    with open(name, 'w') as file:
        json.dump(r.json(), file)
```

```python
def load_json(name):
    with open(name) as file:
        data = json.load(file)
        return data
```

```python
url1 = 'https://datausa.io/api/data?drilldowns=Nation&measures=Population'
api1 = requests_get(url1)
```

```python
dump_json('usa', api1)
```

```python
data1 = load_json('usa')
```

```python
df1 = pd.DataFrame(data1['data'])
df1
```

```python
df1 = df1.drop(df1.columns[2], axis=1)
df1
```

```python
df1 = df1.reindex(df1.index[::-1])
df1
```

```python
x = df1.Year
y = df1.Population
plt.plot(x, y)
plt.title('Population by Year')
plt.xlabel('Year')
plt.ylabel('Population')
plt.show()
```

```python
url3 = 'https://www.7timer.info/bin/astro.php?lon=113.2&lat=23.1&ac=0&unit=metric&output=json&tzshift=0'
api3 = requests_get(url3)
```

```python
dump_json('timer', api3)
```

```python
data3 = load_json('timer')
```

```python
df3 = pd.DataFrame(data3['dataseries'])
df3
```

```python
df3 = df3.drop('wind10m', axis=1)
df3.head()
```

```python
x_rain = df3[df3.prec_type=='rain'].timepoint
y_rain = df3[df3.prec_type=='rain'].temp2m
x_none = df3[df3.prec_type=='none'].timepoint
y_none = df3[df3.prec_type=='none'].temp2m

fig, ax = plt.subplots(1, figsize=(10, 6))
fig.suptitle('temperature by timepoint')

ax.scatter(x_rain, y_rain, color="blue", s=100, alpha=0.5, linewidths=1, label='rain')
ax.scatter(x_none, y_none, color="yellow", s=100, alpha=0.5, linewidths=1, label='not rain')
ax.legend(title='prediction')
ax.grid(True)

plt.show()
```
